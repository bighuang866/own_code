而DataFrame的cov函数则将每一列视为一个序列，并计算出协方差矩阵

```python
frame = pd.DataFrame(np.random.randn(1000, 5), columns=['a', 'b', 'c', 'd', 'e'])
frame
```

```python
frame.cov()
```

DataFrame.cov还支持min_periods参数，该参数指定生成有效结果所需的最小非nan数据数量，如果某列nan太多将无法计算出结果

```python
frame = pd.DataFrame(np.random.randn(20, 3), columns=['a', 'b', 'c'])
```

```python
frame
```

```python
frame.loc[frame.index[:5], 'a'] = np.nan
frame.loc[frame.index[5:10], 'b'] = np.nan
```

```python
frame
```

```python
frame.cov()
```

```python
frame.cov(min_periods=12)
```

## Window Function

在金融时间序列分析中，经常会遇到处理滚动移动平均窗口计算的问题，pandas通过rolling和expanding的机制进行计算  
（如果是计算均线等技术指标，python第三方包talib可以提供更好的计算体验，该包集成了几百种常用及不常用的技术指标计算方式）

```python
s = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000)).cumsum()
```

```python
s
```

```python
r = s.rolling(window=60)
```

```python
r
```

```python
r.mean()
```

```python
s.plot(style='k--')
r.mean().plot(style='k')
```

```python
df = pd.DataFrame(np.random.randn(1000, 4),
                  index=pd.date_range('1/1/2000', periods=1000),
                  columns=['A', 'B', 'C', 'D'])
```

```python
df = df.cumsum()
```

```python
df.rolling(window=60).sum()
```

```python
df.rolling(window=60).sum().plot(subplots=True) # rolling同样支持min_periods参数，可以用来剔除nan太多的计算结果
```

rolling支持大量的内置方法

当然，rolling也提供了apply方法，这给了我们很多自由设计的空间

```python
s1 = pd.Series([1, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9])
s1
```

```python
mad = lambda x: x.iloc[0]+x.iloc[1]-x.iloc[2]+x.iloc[3]+x.iloc[4]
```

```python
s1.rolling(window=5).apply(mad, raw=False)
```

如果对于apply传入的机制不是很清楚，我们可以使用之前使用过的小技巧

```python
def fun(x):
    print(type(x),"\n", x, "\n\n")
    return 0
```

```python
s1.rolling(window=5).apply(fun, raw=False)
```

了解了rolling后，expanding就非常容易理解了

```python
s1.expanding().apply(fun, raw=False)
```

expanding同样支持大量内置函数
